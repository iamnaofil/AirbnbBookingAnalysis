```python
#Replacing those empty values with Mean Median or Mode
mean_df=df['Calories'].mean().round()
mean_df
```

```python
#Replace Data with Wrong Format
#You can See Date COlumn has all values in String
#and a cell having wrong format and a cell having NaT
#Pandas has to_datetime method to convert any string into date of Y-M-d
df['Date']=pd.to_datetime(df['Date'])
print(df.to_string())
```

```python
df['Date'].fillna('2020-12-22',inplace=True)
```

```python
df
```

```python
#To check Data type of Specific Column use dtypes
df['Date'].dtypes
```

```python
#To Drop Na Value from Specific Column
df.dropna(subset=['Date'])
#There is nothing to Drop because we fill it with 2020-12-22 previously
```

```python
#Now you can see 450 value in Duration 7th Index ,
#you can change it to mean of All Duration Column
df.loc[7,'Duration']=df['Duration'].mean().round()
```

```python
df
```

```python
#if have big data you can use for loop for multiple values filling
for x in df.index:
  if df.loc[x,'Duration']>=68:
    df.loc[x,'Duration']=62
df
```

```python
for y in df.index:
  #If any values in Maxpulse is equal to (20+ of Pulse ) Then there is an extra column created having value of that maxpulse +11
  if df.loc[y,'Maxpulse']==(df.loc[y,'Pulse']+20):
    df.loc[y,'OutRange']=df.loc[y,'Maxpulse']+11

```

```python
c_mean=(df['Pulse']+df['Maxpulse']).mean()
df['OutRange']=c_mean
```
